```python
# import numpy
import numpy as np 
```

```python
# Numpy program to create a square matrix
# Dimensions as entered by user
user_dim = input(f"Specify dimensions of matrix, separated by a space:   ")
dim = tuple(int(item)for item in user_dim.split())

# Create matrix
user_dim_matrix = np.ones((dim))
user_dim_matrix = np.zeros((dim), dtype=int)

# Fill matrix with checkerboard pattern
user_dim_matrix[1::2,::2] = 1
user_dim_matrix[::2, 1::2] = 1

```

```python
print(user_dim_matrix)
```

```python
# Numpy program to create a 2D array
# With character X on the border
# And character O inside the array

_2d_array = np.array([["X","X","X","X"], ["X","X","X","X"], ["X","X","X","X"], ["X","X","X","X"]])
_2d_array[1:-1, 1:-1] = "O"
```

```python
# Numpy program to extract all contiguous 3x3 blocks from any 12x12 matrix
# Create original array
orig_arr = np.random.randint(0,4,(12,12))

n = 3
i = 1 + (orig_arr.shape[0]-3)
j = 1 + (orig_arr.shape[1]-3)
outcome = np.lib.stride_tricks.as_strided(orig_arr, shape=(i,j,n,n), strides= orig_arr.strides + orig_arr.strides)
```

```python
print(f"Original array:\n", orig_arr)
```

```python
print("\nContiguous 3x3 blocks:\n")
print("\n", outcome)
```
