```python
# Extract records of all artists 
# Whose names start with the letter "b"
# First remove all NaN values from the database


uch_df = uch_df.dropna()

uch_df.isna().sum()





```

```python
# Extract records of all artists 
# Whose names start with the letter "b"
# Next step using startswith()


uch_df = uch_df[uch_df["artist_mb"].str.startswith("b")]
uch_df
```

```python
# Extract records of all artists 
# Whose names start with the letter "b"
# Next select columns of artist_mb, artist_lastfm to show extracted b names

uch_df[["artist_mb", "artist_lastfm"]]
```

```python
# Add a new column is_USA_based to the DataFrame
# With Yes or No as it's content
# Depending on the country of origin of the artist

uch_df["is_USA_based_mb"] = np.where(
                  uch_df["country_mb"] == "United States",
                  "Yes",
                  "No"
)
```

```python
uch_df
```

```python
# Add a new column is_USA_based to the DataFrame
# With Yes or No as it's content
# Depending on the country of origin of the artist

uch_df["is_USA_based_lastfm"] = np.where(
                  uch_df["country_lastfm"] == "United States",
                  "Yes",
                  "No"
)
```

```python
uch_df
```

```python
# Selecting columns to show artist, country, and artist based in the USA
artist_is_USA_based = uch_df[["artist_mb", "artist_lastfm", "country_mb", "country_lastfm", "is_USA_based_mb", "is_USA_based_lastfm"]]
```
